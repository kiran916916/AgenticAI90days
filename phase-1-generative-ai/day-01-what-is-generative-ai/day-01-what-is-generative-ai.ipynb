## 🧠 Concept
**Discriminative** AI answers *"which label?"* (spam / not-spam). **Generative** AI
answers *"what comes next?"* and produces brand-new content: text, code, images,
audio. Large Language Models (LLMs) like GPT are generative models for text.

```mermaid
flowchart LR
    P["Prompt / context"] --> M["🧠 Generative model"]
    M --> O["New content<br/>(text, code, image, audio)"]
```

The engine under modern GenAI is shockingly simple to state: **predict the next
token, over and over.** A model trained on lots of text learns the statistics of
"what usually follows what," then generates by sampling the next token, appending
it, and repeating.

Today you'll build the *toy* version of that idea: a **Markov text generator**. It
learns "after these characters, which character tends to come next?" and then
babbles new text. It's not intelligent — but it's exactly the right mental model
for "learn patterns → generate."

## 🔒 Solution
*Try it yourself first!* Run the cell below only when you're ready to compare.

````python
import random


def build_model(text, k=2):
    model = {}
    for i in range(len(text) - k):
        prefix = text[i:i + k]
        model.setdefault(prefix, []).append(text[i + k])
    return model


def generate(model, seed, n=200, k=2):
    out = seed
    for _ in range(n):
        choices = model.get(out[-k:])
        if not choices:
            break
        out += random.choice(choices)
    return out


CORPUS = "the cat sat on the mat. the cat ate the rat. the rat ran from the cat. " * 30

if __name__ == "__main__":
    model = build_model(CORPUS)
    print(generate(model, "th", n=180))
````
