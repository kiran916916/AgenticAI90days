## 🧠 Concept
A chat LLM takes a list of **messages** and returns the next message. Messages have
**roles**:
- `system` — standing instructions ("who the assistant is")
- `user` — what the human says
- `assistant` — what the model replies

```mermaid
flowchart LR
    sys["system: 'You are concise'"] --> M["🧠 LLM"]
    usr["user: 'What is GenAI?'"] --> M --> rep["assistant: 'Generative AI…'"]
```

The course wraps all providers (OpenAI, Azure, Ollama) behind one function so you
always call the model the same way:

```python
from shared.llm import chat
chat([{"role": "user", "content": "Hi!"}])
```
