## 🧠 Concept
**Few-shot prompting** = show the model a handful of `input → output` examples
*before* the real input. It copies the pattern. Powerful for classification,
formatting, and style.

Patterns worth knowing:
- **Zero-shot**: just instructions.
- **Few-shot**: instructions + examples.
- **Chain-of-thought** (teaser): "think step by step" for reasoning tasks.
- **Delimiters**: wrap user text in quotes/tags so the model knows what's data.

## 🔒 Solution
*Try it yourself first!* Run the cell below only when you're ready to compare.

```python
from shared.llm import chat


def few_shot_classify(text):
    messages = [
        {"role": "system", "content": "Classify the sentiment as POSITIVE or NEGATIVE. Reply with exactly one word."},
        {"role": "user", "content": "I love this, it's amazing!"},
        {"role": "assistant", "content": "POSITIVE"},
        {"role": "user", "content": "This is terrible and broke instantly."},
        {"role": "assistant", "content": "NEGATIVE"},
        {"role": "user", "content": text},
    ]
    return chat(messages, temperature=0).strip()


if __name__ == "__main__":
    for s in ["This made my whole day!", "Worst purchase ever.", "It's fine, I guess."]:
        print(f"{s!r:30} -> {few_shot_classify(s)}")
```
