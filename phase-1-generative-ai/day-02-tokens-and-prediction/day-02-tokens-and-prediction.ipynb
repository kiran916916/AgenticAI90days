## 🧠 Concept
LLMs don't see letters or words — they see **tokens**: chunks of text (often
sub-words). "tokenization" is the rule that splits text into tokens. Roughly,
1 token ≈ 4 characters of English, or ~¾ of a word.

Why you should care:
- **Cost & limits** are measured in tokens. The **context window** (e.g. 128k
  tokens) is the max the model can attend to at once.
- **Quirks** come from tokens: odd spelling/number behavior often traces back to
  how text was tokenized.

```mermaid

```

Generation = **next-token prediction**: given the tokens so far, the model outputs
a probability for *every* token in its vocabulary; we pick one, append, repeat.

You'll model this with words (not sub-words) to keep it simple: count which word
follows which, then predict the most likely next word.

## 🔒 Solution
*Try it yourself first!* Run the cell below only when you're ready to compare.

````python
import re
from collections import Counter


def tokenize(text):
    return re.findall(r"[a-z']+", text.lower())


def next_word_table(tokens):
    table = {}
    for a, b in zip(tokens, tokens[1:]):
        table.setdefault(a, Counter())[b] += 1
    return table


def predict_next(table, word):
    counter = table.get(word)
    return counter.most_common(1)[0][0] if counter else None


CORPUS = (
    "the quick brown fox jumps over the lazy dog. "
    "the quick brown cat sleeps on the warm mat. "
    "the lazy dog sleeps in the warm sun. "
) * 5

if __name__ == "__main__":
    toks = tokenize(CORPUS)
    print(f"{len(toks)} tokens, {len(set(toks))} unique")
    table = next_word_table(toks)
    for w in ["the", "quick", "lazy", "warm"]:
        print(f"after {w!r:8} -> {predict_next(table, w)!r}")
````
