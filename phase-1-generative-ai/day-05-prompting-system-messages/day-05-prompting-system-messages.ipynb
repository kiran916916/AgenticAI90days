## 🧠 Concept
The system prompt is your model's "constitution": role, tone, rules, output format.
For agents later, this is where the persona, the tool catalog, and the guardrails
all live.

```mermaid
flowchart TD
    SP["📜 System prompt<br/>role · rules · format"] --> B["🧠 LLM"]
    U["🙋 User input"] --> B --> Out["✅ On-spec output"]
```

Rule of thumb: **be explicit about the format you want.** Vague prompt → vague output.

## 🔒 Solution
*Try it yourself first!* Run the cell below only when you're ready to compare.

````python
from shared.llm import chat


def ask_with_persona(user_msg, persona):
    messages = [
        {"role": "system", "content": persona},
        {"role": "user", "content": user_msg},
    ]
    return chat(messages)


if __name__ == "__main__":
    q = "Explain what a variable is."
    print("— pirate —\n", ask_with_persona(q, "You are a cheerful pirate."))
    print("\n— engineer —\n", ask_with_persona(q, "You are a terse senior engineer. Max 20 words."))
````
