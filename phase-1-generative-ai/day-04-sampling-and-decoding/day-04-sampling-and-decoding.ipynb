## 🎯 Objective
Control *how* a model turns next-token probabilities into actual text, and know
which settings to use for reliable vs creative output.

## 🧠 Concept
At each step the model has a probability for every possible next token. **Decoding**
chooses one:
- **temperature** (0–~1.5): low = focused/deterministic; high = random/creative.
- **top-p** (nucleus): sample only from the smallest set of tokens whose
  probabilities sum to *p*.
- **max tokens**: a hard cap on output length (a real cost lever).

Key rule you'll reuse all course long:
> For **decisions and tool use**, keep temperature **low** (≈0) so behavior is
> consistent and parseable. For **brainstorming**, raise it.

```python
from shared.llm import chat

messages = [
    {"role": "system", "content": "You are a poet."},
    {"role": "user", "content": "Describe the ocean in one short line."},
]


def main():
    print("temp 0.0 :", chat(messages, temperature=0.0))
    print("temp 1.0a:", chat(messages, temperature=1.0))
    print("temp 1.0b:", chat(messages, temperature=1.0))
    print("capped   :", chat(messages, temperature=1.0, max_tokens=6))


if __name__ == "__main__":
    main()
```
