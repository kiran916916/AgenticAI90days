````python
from agents import Agent, Runner

spanish = Agent(name="Spanish", instructions="You respond only in Spanish.")
english = Agent(name="English", instructions="You respond only in English.")

triage = Agent(
    name="Triage",
    instructions="Hand off to the agent matching the user's language.",
    handoffs=[spanish, english],
)

result = Runner.run_sync(triage, "Hola, ¿puedes ayudarme?")
print(result.final_output)
````
