## 🔒 Solution

Idiomatic, current-API version.

````python
from agents import Agent, Runner, function_tool

@function_tool
def add(a: int, b: int) -> int:
    return a + b

agent = Agent(
    name="Math helper",
    instructions="Use your tools for any arithmetic.",
    tools=[add],
)

result = Runner.run_sync(agent, "What is 21 + 21?")
print(result.final_output)
````
