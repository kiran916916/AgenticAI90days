## 🔒 Solution

Idiomatic, current-API version.

````python
from typing import TypedDict
from langgraph.graph import StateGraph, START, END

class State(TypedDict):
    topic: str
    joke: str

def make_joke(state: State) -> dict:
    return {"joke": f"Why did the {state['topic']} cross the road? To get to the other side."}

builder = StateGraph(State)
builder.add_node("make_joke", make_joke)
builder.add_edge(START, "make_joke")
builder.add_edge("make_joke", END)

graph = builder.compile()
print(graph.invoke({"topic": "chicken"})["joke"])
````
