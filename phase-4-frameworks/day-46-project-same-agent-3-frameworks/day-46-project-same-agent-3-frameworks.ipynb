## 🔒 Solution

Idiomatic, current-API version.

```python
from shared.llm import chat

def get_weather(city: str) -> str:
    return f"Weather in {city}: 21C, clear."

def weather_agent(question):
    city = chat([
        {"role": "system", "content": "Reply with ONLY the city name mentioned, nothing else."},
        {"role": "user", "content": question},
    ], temperature=0).strip()
    return get_weather(city)

print(weather_agent("What's the weather like in Tokyo today?"))

# The framework versions (Agents SDK, LangGraph) express the SAME three moves:
# 1) read the city, 2) call the weather tool, 3) return the result.
```
