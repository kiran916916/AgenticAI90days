## 🔒 Solution

Verified against autogen-agentchat 0.7.5.

```python
from autogen_agentchat.agents import AssistantAgent
from autogen_agentchat.teams import SelectorGroupChat
from autogen_agentchat.conditions import MaxMessageTermination
from autogen_ext.models.openai import OpenAIChatCompletionClient

mc = OpenAIChatCompletionClient(model="gpt-4o-mini")
math_agent = AssistantAgent("math", model_client=mc, system_message="You handle arithmetic only.")
word_agent = AssistantAgent("words", model_client=mc, system_message="You handle wording/writing only.")

team = SelectorGroupChat([math_agent, word_agent], model_client=mc,
                         termination_condition=MaxMessageTermination(3))
result = await team.run(task="Rewrite 'hi there' more formally.")
for m in result.messages:
    print(f"{m.source:6}: {m.content}")
await mc.close()
```
