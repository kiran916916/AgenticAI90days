## 🔒 Solution

```python
import hashlib, math

def cheap_embed(text, dim=64):
    vec = [0.0] * dim
    for tok in text.lower().split():
        vec[int(hashlib.md5(tok.encode()).hexdigest(), 16) % dim] += 1.0
    norm = math.sqrt(sum(x * x for x in vec)) or 1.0
    return [x / norm for x in vec]

def cosine(a, b):
    return sum(x * y for x, y in zip(a, b))

DOCS = {"a": "Refunds take 5 business days.", "b": "Support hours are 9am-5pm.",
        "c": "Shipping is free over $50."}
EMB = {k: cheap_embed(v) for k, v in DOCS.items()}

def answer_with_citations(query, k=2):
    qe = cheap_embed(query)
    hits = sorted(((cosine(qe, e), id) for id, e in EMB.items()), reverse=True)[:k]
    return f"{DOCS[hits[0][1]]}  [sources: {[id for _, id in hits]}]"

print(answer_with_citations("how long do refunds take"))
```
