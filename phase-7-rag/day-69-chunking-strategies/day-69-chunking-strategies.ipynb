## 🔒 Solution

```python
def chunk(text, size=40, overlap=10):
    chunks = []
    step = size - overlap
    for i in range(0, len(text), step):
        piece = text[i:i + size]
        if piece:
            chunks.append(piece)
        if i + size >= len(text):
            break
    return chunks

para = "Agents observe, decide, act, and repeat. " * 3
for c in chunk(para):
    print(repr(c))
```
