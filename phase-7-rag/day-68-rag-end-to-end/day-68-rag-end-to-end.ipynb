## 🔒 Solution

```python
DOCS = {
    "d1": "The Eiffel Tower is in Paris, France.",
    "d2": "Mount Fuji is the tallest mountain in Japan.",
    "d3": "The Colosseum is an ancient arena in Rome, Italy.",
}

def retrieve(query):
    qw = set(query.lower().split())
    scored = [(len(qw & set(text.lower().split())), key, text) for key, text in DOCS.items()]
    scored.sort(reverse=True)
    return scored[0]

def answer(query):
    overlap, key, text = retrieve(query)
    return f"{text}  (source: {key})"

print(answer("Where is the Eiffel Tower?"))
```
