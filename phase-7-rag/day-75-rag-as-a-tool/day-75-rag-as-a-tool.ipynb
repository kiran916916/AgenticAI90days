## 🔒 Solution

```python
import hashlib, math

def cheap_embed(text, dim=64):
    vec = [0.0] * dim
    for tok in text.lower().split():
        vec[int(hashlib.md5(tok.encode()).hexdigest(), 16) % dim] += 1.0
    norm = math.sqrt(sum(x * x for x in vec)) or 1.0
    return [x / norm for x in vec]

def cosine(a, b):
    return sum(x * y for x, y in zip(a, b))

DOCS = {"d1": "Our API rate limit is 100 requests per minute.",
        "d2": "The CEO is Dana Smith.", "d3": "We were founded in 2015."}
EMB = {k: cheap_embed(v) for k, v in DOCS.items()}

def search_docs(query):
    qe = cheap_embed(query)
    best = max(EMB.items(), key=lambda kv: cosine(qe, kv[1]))
    return DOCS[best[0]]

TOOLS = {"search_docs": search_docs}

def use_tool(name, arg):
    fn = TOOLS.get(name)
    return fn(arg) if fn else f"unknown tool: {name}"

print(use_tool("search_docs", "what is the rate limit"))
```
