## 🔒 Solution

```python
def precision_recall(retrieved, relevant):
    r, rel = set(retrieved), set(relevant)
    tp = len(r & rel)
    precision = tp / len(r) if r else 0.0
    recall = tp / len(rel) if rel else 0.0
    return precision, recall

p, rec = precision_recall(["d1", "d2", "d9"], ["d1", "d2", "d3"])
print(f"precision {p:.2f}  recall {rec:.2f}")
```
