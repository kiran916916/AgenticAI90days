## 🔒 Solution

```python
class Blackboard:
    def __init__(self):
        self.data = {}
    def write(self, key, value):
        self.data[key] = value
    def read(self, key, default=None):
        return self.data.get(key, default)

bb = Blackboard()

def collector(bb):
    bb.write("facts", ["sky is blue", "water is wet"])

def summarizer(bb):
    facts = bb.read("facts", [])
    bb.write("summary", f"{len(facts)} facts: " + "; ".join(facts))

collector(bb); summarizer(bb)
print(bb.read("summary"))
```
