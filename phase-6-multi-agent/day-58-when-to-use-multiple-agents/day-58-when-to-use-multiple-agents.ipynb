## 🔒 Solution

```python
def team_advice(task: str) -> str:
    seams = task.lower().count(" and ") + task.lower().count(" then ")
    return "team" if seams >= 2 else "single agent"

for t in ["Summarize this", "Research X and draft an email and translate it"]:
    print(t, "->", team_advice(t))
```
