## 🔒 Solution

```python
def proposer(question):
    return "The answer is 42 because it is the classic placeholder."

def critic(claim):
    return "weak: no evidence" if "because" not in claim else "ok: has a reason"

def judge(claim, critique):
    return "accept" if critique.startswith("ok") else "revise"

c = proposer("why 42?"); cr = critic(c)
print(cr, "->", judge(c, cr))
```
