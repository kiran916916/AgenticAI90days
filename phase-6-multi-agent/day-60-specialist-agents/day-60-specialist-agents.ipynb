## 🔒 Solution

```python
import re
from shared.tools import calculator

def math_specialist(task):
    exprs = re.findall(r"[0-9\.\+\-\*\/\(\) ]{3,}", task)
    return calculator(max(exprs, key=len).strip()) if exprs else "no expression found"

def writing_specialist(task):
    return f"DRAFT: {task.strip().capitalize()}."

SPECIALISTS = {"math": math_specialist, "writing": writing_specialist}

def route(task):
    t = task.lower()
    return "math" if any(w in t for w in ["+", "*", "sum", "add"]) else "writing"

def supervisor(task):
    name = route(task)
    specialist = SPECIALISTS.get(name, lambda t: "no specialist")
    return f"[{name}] {specialist(task)}"

print(supervisor("what is 6 * 7"))
print(supervisor("write a welcome line"))
```
