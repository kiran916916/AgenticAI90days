## 🔒 Solution

```python
from shared.evals import GoldenCase, run_eval

def keyword_judge(q, a, criteria):
    kws = [k.strip().lower() for k in criteria.split(",")]
    hit = sum(1 for k in kws if k in a.lower())
    return hit / len(kws), f"{hit}/{len(kws)}"

single = lambda q: "Paris"
team   = lambda q: "The capital of France is Paris, a major European city."
cases = [GoldenCase("fr", "capital of France", "paris, city")]

print("single agent:"); r1 = run_eval(single, cases, judge=keyword_judge)
print("\nteam:");        r2 = run_eval(team, cases, judge=keyword_judge)
print("\nteam better?", r2[0]["score"] > r1[0]["score"])
```
