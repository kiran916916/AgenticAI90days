## 🔒 Solution

```python
def route(task: str) -> str:
    t = task.lower()
    if any(w in t for w in ["add", "plus", "sum", "calculate", "+", "*"]):
        return "math"
    if any(w in t for w in ["write", "slogan", "email", "rewrite"]):
        return "writing"
    if "weather" in t:
        return "weather"
    return "general"

for t in ["add 2 and 3", "write a slogan", "weather in Rome", "hello"]:
    print(f"{t:18} -> {route(t)}")
```
