## 🔒 Solution

```python
class Budget:
    def __init__(self, max_steps=5):
        self.max_steps = max_steps
        self.steps = 0
    def can_continue(self):
        return self.steps < self.max_steps
    def spend(self, n=1):
        self.steps += n

def run_team(budget):
    turns = []
    while budget.can_continue():
        budget.spend()
        turns.append(f"turn {budget.steps}")
        if budget.steps == 3:    # goal reached early
            break
    return turns

print(run_team(Budget(max_steps=10)))
```
