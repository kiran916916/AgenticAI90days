# Day 61 — Handoffs & delegation

A handoff passes **work + context** from one agent to the next. The key is carrying enough
context that the receiver can continue without redoing work. > ✅ Offline.

```python
# ▶ Run this first — puts the repo root on sys.path so `shared/` imports work.
import sys, pathlib
root = pathlib.Path.cwd()
while root != root.parent and not (root / "shared" / "llm.py").exists():
    root = root.parent
if str(root) not in sys.path:
    sys.path.insert(0, str(root))
print("repo root on sys.path:", root)
```

## 🔒 Solution

```python
def researcher(topic):
    return f"notes on {topic}: founded 2015; mission is X; 3 products"

def writer(context):
    return f"Article based on -> {context}"

def pipeline(topic):
    notes = researcher(topic)
    return writer(notes)

print(pipeline("Acme Corp"))
```
