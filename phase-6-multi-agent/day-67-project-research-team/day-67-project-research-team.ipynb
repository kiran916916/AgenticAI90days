## 🔒 Solution

```python
from shared.llm import chat

class Blackboard:
    def __init__(self): self.data = {}
    def write(self, k, v): self.data[k] = v
    def read(self, k, d=None): return self.data.get(k, d)

def researcher(question, bb):
    notes = chat([{"role": "system", "content": "List 3 short factual bullet points."},
                  {"role": "user", "content": question}], temperature=0)
    bb.write("notes", notes)

def writer(question, bb):
    return chat([{"role": "system", "content": "Write a 2-sentence answer using the notes."},
                 {"role": "user", "content": f"Q: {question}\nNotes:\n{bb.read('notes')}"}], temperature=0)

def research_team(question):
    bb = Blackboard()
    researcher(question, bb)
    return writer(question, bb)

print(research_team("Why is the sky blue?"))
```
