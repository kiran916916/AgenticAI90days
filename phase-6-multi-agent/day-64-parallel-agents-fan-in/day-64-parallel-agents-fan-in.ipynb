## 🔒 Solution

```python
from concurrent.futures import ThreadPoolExecutor

def angle_a(topic): return f"pro: {topic} saves time"
def angle_b(topic): return f"con: {topic} has a learning curve"
def angle_c(topic): return f"cost: {topic} is mid-priced"

def gather(topic, specialists):
    with ThreadPoolExecutor() as ex:
        futures = {name: ex.submit(fn, topic) for name, fn in specialists.items()}
        return {name: f.result() for name, f in futures.items()}

results = gather("the tool", {"a": angle_a, "b": angle_b, "c": angle_c})
for name, text in results.items():
    print(name, "->", text)
```
