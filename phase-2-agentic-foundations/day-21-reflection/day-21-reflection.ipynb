## 🔒 Solution

One correct way to do it. Compare it with your version.

```python
from shared.llm import chat

def draft(goal):
    return chat([{"role": "user", "content": goal}], temperature=0.5)

def critique(goal, answer):
    return chat([
        {"role": "system", "content":
            "You are a strict reviewer. If the answer fully and correctly addresses the goal, "
            "reply exactly 'OK'. Otherwise reply with a short, concrete instruction to fix it."},
        {"role": "user", "content": f"Goal: {goal}\n\nAnswer: {answer}"},
    ], temperature=0)

def answer_with_reflection(goal):
    a = draft(goal)
    c = critique(goal, a)
    if c.strip().upper().startswith("OK"):
        return a
    return chat([
        {"role": "system", "content": "Revise the answer to satisfy the reviewer. Return only the improved answer."},
        {"role": "user", "content": f"Goal: {goal}\n\nDraft: {a}\n\nReviewer says: {c}"},
    ], temperature=0)

print(answer_with_reflection("List 3 risks of giving an agent shell access. Be specific."))
```
