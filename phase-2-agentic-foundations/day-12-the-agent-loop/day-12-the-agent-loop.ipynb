## 🔒 Solution

````python
import random


class NumberWorld:
    def __init__(self, low=1, high=100):
        self.secret = random.randint(low, high)

    def react(self, guess):
        if guess == self.secret:
            return "correct"
        return "higher" if guess < self.secret else "lower"


def choose_guess(low, high):
    return (low + high) // 2


def update_state(low, high, guess, obs):
    if obs == "higher":
        low = guess + 1
    elif obs == "lower":
        high = guess - 1
    return low, high


def main():
    world, low, high = NumberWorld(), 1, 100
    for turn in range(1, 11):
        guess = choose_guess(low, high)
        obs = world.react(guess)
        print(f"turn {turn:2d}: range=[{low:3d},{high:3d}] guess={guess:3d} -> {obs}")
        if obs == "correct":
            print(f"✅ Solved in {turn} turns!"); return
        low, high = update_state(low, high, guess, obs)
    print("❌ Out of turns.")


if __name__ == "__main__":
    main()
````
