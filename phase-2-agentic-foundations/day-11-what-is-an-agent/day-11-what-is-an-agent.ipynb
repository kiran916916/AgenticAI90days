## 🎯 Objective
See what turns a generative model into an **agent**, and feel it by building the

## 🧠 Concept
A model on its own just *produces text*. An **agent** wraps a brain (rules or an
LLM) in a **loop** with **autonomy**: it observes, decides on an action, *acts on
the world*, sees the result, and repeats — until a goal is met.

```mermaid
flowchart LR
    A["Single prompt<br/>(ask once)"] --> B["Chain<br/>(fixed steps)"] --> C["Agent<br/>(chooses steps in a loop,<br/>uses tools)"]
```

Today's "brain" is just an `if`/`else` — a **reflex agent** that maps the current
observation straight to an action. Master the *structure* first; we add the LLM
brain back as the decision-maker soon.

````python
from dataclasses import dataclass


@dataclass
class Room:
    temperature: float

    def step(self, action):
        if action == "HEAT":
            self.temperature += 1.0
        elif action == "COOL":
            self.temperature -= 1.0
        else:
            self.temperature += (18.0 - self.temperature) * 0.1


TARGET, TOLERANCE = 21.0, 0.5


def decide_action(temp):
    if temp < TARGET - TOLERANCE:
        return "HEAT"
    if temp > TARGET + TOLERANCE:
        return "COOL"
    return "OFF"


def main():
    room = Room(15.0)
    for tick in range(1, 16):
        action = decide_action(room.temperature)
        print(f"tick {tick:2d} | {room.temperature:5.1f}°C | {action}")
        room.step(action)


if __name__ == "__main__":
    main()
````
