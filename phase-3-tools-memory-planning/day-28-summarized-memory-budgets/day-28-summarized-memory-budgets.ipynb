## 🔒 Solution

One correct way to do it.

```python
def char_len(messages):
    return sum(len(m["content"]) for m in messages)

def trim_to_budget(messages, max_chars=120):
    system, rest = messages[0], messages[1:]
    kept_reversed = []
    used = len(system["content"])
    for m in reversed(rest):
        if used + len(m["content"]) > max_chars:
            break
        kept_reversed.append(m)
        used += len(m["content"])
    return [system] + list(reversed(kept_reversed))

msgs = [{"role": "system", "content": "S" * 20}] + \
       [{"role": "user", "content": f"message-{i}"} for i in range(10)]
kept = trim_to_budget(msgs, max_chars=120)
print("kept", len(kept), "messages,", char_len(kept), "chars")
print([m["content"] for m in kept])
```
