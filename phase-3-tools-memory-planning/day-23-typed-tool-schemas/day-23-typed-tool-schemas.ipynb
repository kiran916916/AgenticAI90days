## 🔒 Solution

One correct way to do it.

```python
import json
from pydantic import BaseModel, Field, ValidationError

class WeatherArgs(BaseModel):
    city: str = Field(..., min_length=1)
    units: str = Field("celsius")

def get_weather(args: dict):
    try:
        a = WeatherArgs(**args)
    except ValidationError as exc:
        e = exc.errors()[0]
        return f"Error: {e['msg']} (field: {e['loc']})"
    return f"Weather in {a.city}: 21 {a.units}"

print(get_weather({"city": "Seattle"}))
print(get_weather({"units": "f"}))
print("\nSchema the model sees:")
print(json.dumps(WeatherArgs.model_json_schema(), indent=2))
```
