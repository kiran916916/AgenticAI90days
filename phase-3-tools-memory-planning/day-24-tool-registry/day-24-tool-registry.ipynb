## 🔒 Solution

One correct way to do it.

```python
from shared.tools import calculator, clock, word_count

class Registry:
    def __init__(self):
        self._tools = {}
    def register(self, name, fn):
        self._tools[name] = fn
    def names(self):
        return list(self._tools)
    def call(self, name, arg):
        fn = self._tools.get(name)
        if fn is None:
            return f"unknown tool: {name} (have: {', '.join(self.names())})"
        try:
            return str(fn(arg))
        except Exception as exc:
            return f"Error running {name}: {exc}"

r = Registry()
for n, f in [("calc", calculator), ("clock", clock), ("words", word_count)]:
    r.register(n, f)
print(r.names())
print(r.call("calc", "2+2"))
print(r.call("nope", "x"))
```
