## 🔒 Solution

One correct way to do it.

```python
import time
from shared.tracing import Tracer
from shared.tools import calculator

tr = Tracer()

def slow_lookup(x):
    time.sleep(0.03)
    return f"looked up {x}"

def run():
    with tr.span("calculator"):
        calculator("2+2")
    with tr.span("slow_lookup", arg="Seattle"):
        slow_lookup("Seattle")
    tr.print_trace()

run()
```
