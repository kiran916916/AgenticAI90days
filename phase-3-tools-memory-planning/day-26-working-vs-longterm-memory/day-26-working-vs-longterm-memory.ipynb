- **Working memory** — a scratchpad for the *current* task; cleared when it's done.

## 🔒 Solution

One correct way to do it.

```python
class Memory:
    def __init__(self):
        self.working = []
        self.longterm = {}
    def note(self, text):
        self.working.append(text)
    def remember(self, key, value):
        self.longterm[key] = value
    def recall(self, key):
        return self.longterm.get(key, "(unknown)")
    def end_task(self):
        self.working = []     # scratchpad gone; facts stay

m = Memory()
m.note("computed 2+2"); m.remember("name", "Sam")
print("during task:", m.working, "| name:", m.recall("name"))
m.end_task()
print("after task: ", m.working, "| name:", m.recall("name"))
```
