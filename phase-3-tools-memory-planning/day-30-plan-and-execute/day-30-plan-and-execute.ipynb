## 🔒 Solution

One correct way to do it.

```python
from shared.llm import chat
from shared.tools import calculator

def do_step(step, context):
    raw = chat([
        {"role": "system", "content":
            "Complete ONE step. If arithmetic is needed reply exactly 'CALC: <expression>'. "
            "Otherwise reply with the result text only."},
        {"role": "user", "content": f"Context so far:\n{context}\n\nStep: {step}"},
    ], temperature=0).strip()
    if raw.upper().startswith("CALC:"):
        return calculator(raw.split(":", 1)[1].strip())
    return raw

def run_plan(steps):
    context, out = "", []
    for step in steps:
        result = do_step(step, context)
        out.append((step, result))
        context += f"\n{step} => {result}"
    return out

for step, result in run_plan(["Compute 12 * 12", "State that result in a friendly sentence"]):
    print("-", step, "=>", result)
```
