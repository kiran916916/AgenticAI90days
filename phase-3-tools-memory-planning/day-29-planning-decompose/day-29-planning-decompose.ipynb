## 🔒 Solution

One correct way to do it.

````python
import json
from shared.llm import chat

def plan(goal):
    raw = chat([
        {"role": "system", "content":
            "Return ONLY a JSON array of 2-5 short imperative step strings to achieve the goal."},
        {"role": "user", "content": goal},
    ], temperature=0).strip()
    if raw.startswith("```"):
        raw = raw.strip("`").split("\n", 1)[-1]
    steps = json.loads(raw)
    return [str(s) for s in steps]

for i, step in enumerate(plan("Plan a small birthday party"), 1):
    print(i, step)
````
