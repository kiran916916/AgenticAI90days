## 🔒 Solution

```python
from shared.evals import GoldenCase, run_eval, regression_gate

def keyword_judge(q, a, criteria):
    kws = [k.strip().lower() for k in criteria.split(",")]
    hit = sum(1 for k in kws if k in a.lower())
    return hit / len(kws), f"{hit}/{len(kws)}"

def my_agent(q):
    return {"2+2": "4", "capital of france": "Paris"}.get(q.lower(), "?")

cases = [GoldenCase("math", "2+2", "4"), GoldenCase("geo", "capital of France", "paris")]
results = run_eval(my_agent, cases, judge=keyword_judge)
regression_gate(results, threshold=0.8)
```
