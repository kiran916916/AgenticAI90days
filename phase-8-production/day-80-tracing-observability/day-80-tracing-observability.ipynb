## 🔒 Solution

```python
import json, time
from shared.tracing import Tracer

tr = Tracer()
with tr.span("retrieve", docs=3):
    time.sleep(0.01)
with tr.span("generate", tokens=120):
    time.sleep(0.01)

def export(tracer):
    return json.dumps(
        [{"name": s.name, "duration_ms": s.duration_ms, **s.attributes} for s in tracer.spans],
        indent=2,
    )

print(export(tr))
```
