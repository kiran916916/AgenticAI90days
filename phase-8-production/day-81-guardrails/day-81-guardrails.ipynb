## 🔒 Solution

```python
BLOCKED = ["password", "ssn", "credit card number"]

def input_guard(text):
    low = text.lower()
    for term in BLOCKED:
        if term in low:
            return False, f"blocked: {term}"
    return True, "ok"

def output_guard(text):
    if not text.strip():
        return False, "empty output"
    if len(text) > 500:
        return False, "output too long"
    return True, "ok"

print(input_guard("what is my SSN"))
print(input_guard("what is the weather"))
print(output_guard("a normal answer"))
```
