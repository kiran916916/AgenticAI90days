## 🔒 Solution

```python
from shared.evals import GoldenCase, run_eval, regression_gate

def keyword_judge(q, a, criteria):
    kws = [k.strip().lower() for k in criteria.split(",")]
    hit = sum(1 for k in kws if k in a.lower())
    return hit / len(kws), f"{hit}/{len(kws)}"

def agent(q):
    return {"hi": "Hello there!"}.get(q.lower(), "?")

def test_agent_quality():
    cases = [GoldenCase("greet", "hi", "hello")]
    results = run_eval(agent, cases, judge=keyword_judge)
    assert regression_gate(results, threshold=0.8), "quality regressed!"

test_agent_quality()
print("CI gate passed ✅")
```
