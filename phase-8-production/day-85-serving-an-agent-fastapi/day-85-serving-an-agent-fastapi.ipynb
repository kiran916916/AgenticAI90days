## 🔒 Solution

````python
from fastapi import FastAPI
from pydantic import BaseModel

app = FastAPI()

class Query(BaseModel):
    question: str

@app.post("/chat")
def chat_endpoint(q: Query):
    return {"answer": f"echo: {q.question}"}

routes = [(r.path, sorted(getattr(r, "methods", []) or [])) for r in app.routes]
for path, methods in routes:
    if "POST" in methods:
        print("POST", path)
print("\nServe it:  uvicorn app:app --reload")
````
